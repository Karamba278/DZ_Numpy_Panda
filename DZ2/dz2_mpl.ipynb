```python
# Задание 2
# С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.
t = np.linspace(0, 10, num = 51)
t
```

```python
# Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.
f = np.cos(t)
f
```

```python
# Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. 
# Линия графика должна быть зеленого цвета.


plt.plot(t, f, color='green') 

plt.show() 
```

```python
# Выведите название диаграммы - 'График f(t)'. 
# Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.
# Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.
plt.plot(t, f, color='green') 

plt.title("График f(t)")

plt.xlabel("Значения t")

plt.ylabel("Значения f")

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.grid()

plt.show() 
```

```python
# Задание 3
# С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
x = np.linspace(-3, 3, num = 51)
x
```

```python
#Создайте массивы y1, y2, y3, y4 по следующим формулам:
#y1 = x**2
#y2 = 2 * x + 0.5
#y3 = -3 * x - 1.5
#y4 = sin(x)

y1 = x**2

y2 = 2 * x + 0.5

y3 = -3 * x - 1.5

y4 = np.sin(x)

print('y1= ', y1)

print('y2= ', y2)

print('y3= ', y3)

print('y4= ', y4)
```

```python
# Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure 
# с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, 
# состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.
# В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, 
# в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.
# Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
# Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
# Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.


fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8.5, 3.5)

fig.subplots_adjust(wspace=0.3, hspace=0.6) # Вертикальный зазор сделал побольше, т.к. тогда надписи налезают на графики

ax = ax.flatten()

ax[0].plot(x, y1)

ax[0].axis([-5, 5, 0, 10])

ax[0].set_title("График y1")

ax[1].plot(x, y2)

ax[1].set_title("График y2")

ax[2].plot(x, y3)

ax[2].set_title("График y3")

ax[3].plot(x, y4)

ax[3].set_title("График y4")
```

```python
# 9 Создать одномерный массив а, состоящий из последовательности целых чисел от 0 до 12. 
# Поменять форму этого массива, чтобы получилась матрица A (двумерный массив Numpy), состоящая из 4 строк и 3 столбцов. 
# Получить матрицу At путем транспонирования матрицы A. 
# Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения. 
# Какой размер имеет матрица B? Получится ли вычислить обратную матрицу для матрицы B и почему?

a9 = np.arange(0, 12, 1)
a9
```

```python
a9 = a9.reshape(4,3)
a9
```

```python
a9_t = a9.T
a9_t
```

```python
b9 = a9.dot(a9_t)
b9
```

```python
b9.size # a9 - 4x3, a9_t - 3x4, Следовательно размер матрицы b - 4x4, и, соответственно, size 4*4=16
```

```python
b9_inv = np.linalg.inv(b9)
b9_inv
```

```python
# не получается вычислить обратную матрицу. Проверим почему:

d = np.linalg.det(b9) # Найдем определитель

d
```

```python
# Определитель = 0, значит обратную матрицу посчитать действительно нельзя. 
# Определитель = 0, т.к.: определители матрицы и транспонированной матрицы равны (да, они не квадратные, и опредетели у них это вектора),
# но при замене местами строк со столбцами - знак определителя меняется на противположный, следовательно -x + x = 0
```
